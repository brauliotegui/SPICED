## Creating numpy arrays

```python
data = []
for i in range(1, 10):
    d = i ** 2
    data.append(d)
```

```python
data
```

```python
def square(x):
    return x ** 2
```

```python
data = [square(i) ** 2 for i in range(1, 10)]
data
```

## We can use numpy to make for loops go away

## Two-Dimensional arrays

```python
a = np.zeros((100, 400), dtype=np.uint8)  #  <--- 0...255
a
```

```python
from PIL import Image  # pip install pillow
```

```python
Image.fromarray(a)
```

```python
b = np.random.randint(0, 255, size=(100,400), dtype=np.uint8)
Image.fromarray(b)
```

#### 6. Change the shape

```python
c = b.reshape(200, 200)
c.shape
```

```python
Image.fromarray(c)
```
