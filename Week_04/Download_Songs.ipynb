```python
import requests
from bs4 import BeautifulSoup
```

```python
# Choose the URL you want to request
url = 'https://www.lyrics.com/artist/Pink-Floyd/76669'
```

```python
# Send the request
request = requests.get(url)

# Check the status code in order to assess wether our request was succesful
request.status_code
```

```python
pinkfloyd_songs = request.text
```

```python
type(pinkfloyd_songs)
```

```python
pink_soup = BeautifulSoup(pinkfloyd_songs, 'html.parser')
```

```python
pink_soup.body.find_all(class_='tal qx')
```

```python
pink_soup.find(class_='tal qx').find('a')['href']
```

```python
for td in pink_soup.find_all('td'):
    if "tal" in td.get('class', []):
         print("https://www.lyrics.com/" + (td.find('a')['href']))
```

```python
links = []

for td in pink_soup.find_all('td'):
    if "tal" in td.get('class', []):
         links = (td.find('a')['href'])
```

```python
len(links)
```

```python
#Loop to create all lyric files

for i in range(len(links)):
    temp_url = "http://www.lyrics.com/" + links[i]
    temp_req = requests.get(temp_url)

    with open("pinkfloyd_songs/lyrics/" + str(i) + '.txt',  'w') as file:
        file.write(temp_req.text)
file.close()
```
